```python
import numpy as np
import pandas as pd
import pyspark
from pyspark.ml.feature import Tokenizer, HashingTF, IDF
from pyspark.ml import Pipeline
from pyspark.sql import SparkSession
from tqdm import tqdm
import os
import sys
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder, CrossValidatorModel
import matplotlib.pyplot as plt
from pyspark.ml.feature import Word2Vec


os.environ['PYSPARK_PYTHON'] = sys.executable
os.environ['PYSPARK_DRIVER_PYTHON'] = sys.executable
```

```python
data = pd.read_csv('train.csv')
```

```python
data.head()
```

```python
for row in data.comment_text.values[:5]:
    print(row)
```

```python
spark = SparkSession \
        .builder \
        .appName("pyspark_hw5") \
        .getOrCreate()

spark.sparkContext.setLogLevel("WARN")
```

# Блок 1. HashingTF и IDF

## 1) Подготовить фичи комментариев с помощью HashingTF и IDF

```python
data_spark = spark.createDataFrame(data)
```

```python
tokenizer = Tokenizer(inputCol="comment_text", outputCol="comment_preprocessed")
data_spark = tokenizer.transform(data_spark)
```

```python
feature_num_list = [5, 10, 30, 100, 300, 500, 1000]

for feature_num in tqdm(feature_num_list):
    tf = HashingTF(inputCol="comment_preprocessed", 
                          outputCol="tf_{}".format(feature_num), numFeatures=feature_num)
    idf = IDF(inputCol='tf_{}'.format(feature_num), outputCol="tf_idf_{}".format(feature_num))
    pipe = Pipeline(stages=[tf, idf])
    tf_idf = pipe.fit(data_spark)
    data_spark = tf_idf.transform(data_spark)
```

```python
data_spark.printSchema()
```

```python
data_spark.select(data_spark.columns[:10]).show(5)
```

## 2) Обучить линейные классификаторы и сравнить метрики качества моделей

```python
data.columns[2:].values
```

```python
class_2_feature_num = {}
evaluator = BinaryClassificationEvaluator()

for category in tqdm(data.columns[2:].values):
    feature_num_2_score = {}
    for feature_num in tqdm(feature_num_list):
        feature_name = "tf_idf_{}".format(feature_num)
        subdf = data_spark.select(category, feature_name).withColumnRenamed(category,  "label").\
        withColumnRenamed(feature_name,  "features")
        
        log_reg = LogisticRegression(regParam=0.05, maxIter=8)
        grid = ParamGridBuilder().addGrid(log_reg.regParam, [0.05]).addGrid(log_reg.maxIter, [8]).build()
        cv = CrossValidator(estimator=log_reg, estimatorParamMaps=grid, evaluator=evaluator,  parallelism=4)
        cvModel = cv.fit(subdf)
        predictions = cvModel.transform(subdf)
        score = evaluator.evaluate(predictions)
        
        feature_num_2_score[feature_num] = score
    class_2_feature_num[category] = feature_num_2_score
```

## 3)Сделать выводы о влиянии параметра numFeatures в HashingTF на качество метрик

```python
def plot_metrics(class_2_feature_num, title):
    
    for category in class_2_feature_num.keys():
        x = sorted(class_2_feature_num[category].keys())
        y = [class_2_feature_num[category][key] for key in x]
        plt.plot(x, y, '-o', label=category)

    plt.grid()
    plt.title(title)
    plt.xlabel('feature num')
    plt.ylabel('score')
    plt.legend()
    plt.show()
```

```python
plot_metrics(class_2_feature_num, "tf-idf")
```

С ростом размерности признаков качество растет

# Блок 2. Word2Vec

## 1)Подготовить фичи комментариев с помощью w2v

```python
data_spark = spark.createDataFrame(data)
tokenizer = Tokenizer(inputCol="comment_text", outputCol="comment_preprocessed")
data_spark = tokenizer.transform(data_spark)

feature_num_list = [i for i in range(1, 20, 2)]

for feature_num in tqdm(feature_num_list):
    word2Vec = Word2Vec(vectorSize=feature_num, minCount=5, inputCol="comment_preprocessed",
                        outputCol='word2vec_{}'.format(feature_num))
    model = word2Vec.fit(data_spark)
    data_spark = model.transform(data_spark)
```

## 2)Обучить линейный классификатор и сравнить метрики качества с предыдущими подходами

```python
class_2_feature_num = {}
evaluator = BinaryClassificationEvaluator()

for category in tqdm(data.columns[2:].values):
    feature_num_2_score = {}
    for feature_num in tqdm(feature_num_list):
        feature_name = "word2vec_{}".format(feature_num)
        subdf = data_spark.select(category, feature_name).withColumnRenamed(category,  "label").\
        withColumnRenamed(feature_name,  "features")
        
        log_reg = LogisticRegression(regParam=0.05, maxIter=8)
        grid = ParamGridBuilder().addGrid(log_reg.regParam, [0.05]).addGrid(log_reg.maxIter, [8]).build()
        cv = CrossValidator(estimator=log_reg, estimatorParamMaps=grid, evaluator=evaluator,  parallelism=4)
        cvModel = cv.fit(subdf)
        predictions = cvModel.transform(subdf)
        score = evaluator.evaluate(predictions)
        
        feature_num_2_score[feature_num] = score
    class_2_feature_num[category] = feature_num_2_score
```

```python
plot_metrics(class_2_feature_num, "word2vec")
```
